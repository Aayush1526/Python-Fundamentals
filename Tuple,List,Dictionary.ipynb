```python
#(i) create a Tuple of 10 sports names,do all operations addition,remove,change elements in the Tuple.
```

```python
sports_tup=("shooting","badminton","basketball","cricket","swimming","hockey","golf","football","baseball","table tennis")
#Convert to List(Tuple is imutable)
sports=list(sports_tup)
print(sports)
```

```python
#Adding element
sports.append("cycling")
print(sports)
```

```python
#Removing element
sports.remove("table tennis")
print(sports)
```

```python
#pop element removes last element
sports.pop()
print(sports)
```

```python
#pop element by index
sports.pop(6)
print(sports)
```

```python
#Extend list
sports_new=("horse riding","chess","powerlifting")
sports_new=list(sports_new)
sports.extend(sports_new)
print(sports)
```

```python
#Change the element by index number
sports[4]="Table Tennis"
print(sports) 
```

```python
#Get the Position of element by Index number
print(sports.index("badminton"))
```

```python
#Insert new sports at specific position
sports.insert(0,"Archery")
print(sports)
```

```python
#Sort the element by ascending order
sports.sort()
print(sports)
```

```python
#Sort the element by descending order
sports.sort(reverse=True)
print(sports)
```

```python
#Copy list to another list
sports_new=sports.copy()
print(sports_new)
```

```python
#Reverse the list
sports.reverse()
print(sports)
```

```python
#convert list to tuple
sports=tuple(sports)
print(sports)
```

```python
# (ii). Write a python program to interchange a list items with & without using temporary values.
```

```python
#Without Temporary method
list=["A","B","C",10,20,30,40]
print("List Before interchanging:",list)
idx_1=0
idx_2=3
idx_3=5
list[idx_1],list[idx_2],list[idx_3]=list[idx_3],list[idx_1],list[idx_2]
print("After Interchanging list:",list)
```

```python
# using Temporary value
idx_1 = 1
idx_2 = 4
print("List Before interchanging:",list)
temp = list[idx_1]
list[idx_1] = list[idx_2]
list[idx_2] = temp
print("After Interchanging with temp:",list)
```

```python
# (iii). Create a dictionary, do all operations like addition of elements, remove elements, copy items & create nested dictionaries too.
```

```python
#Add a key:value pair in dictionary
inventory={
    "Laptop":40,
    "Mouse":50,
    "Keyboard":30,
    "Monitor":25,
}
inventory.update({"Display":15})      #update() method
inventory["Graphic Card"]=10          #Assignment method
print(inventory)
```

```python
#Accessing values
print(inventory["Laptop"])        
print(inventory.get("Mouse"))          #.get() method returns none if key dosen't exist
```

```python
#Updating items
print(f"Before Updating {inventory}")
inventory["Mouse"]= 45
print(f"After updating {inventory}")
```

```python
#Removing Item using pop() method
inventory.pop("Graphic Card") 
print(inventory)
```

```python
#popitem() removes the last inserted key-value pair
inventory.popitem()
print(inventory)
```

```python
#Removing Item using del method
inventory={
    "Laptop":40,
    "Mouse":50,
    "Keyboard":30,
    "Monitor":25,
}
del inventory["Monitor"]
print(inventory)
```

```python
#Return the keys of dictionary
print(inventory.keys())
```

```python
#Return the values of dictionary
print(inventory.values())
```

```python
#Return the items of dictionary in key value pair
print(inventory.items())
```

```python
#merge the dictionaries
inventory_2={
    "Laptop Stand":20,
    "Pen Drive":60,
    "Mouse Pad":50,
    "HDMI Cable":20
}
inventory.update(inventory_2)
print(inventory)
```

```python
#Create nested dictionary
countries = {
    "India": {"capital": "New Delhi", "currency": "INR", "states": {"UP": "Lucknow", "MH": "Mumbai"}},
    "USA": {"capital": "Washington D.C.", "currency": "USD", "states": {"CA": "Sacramento", "NY": "Albany"}}
}
print(countries)
```

```python
#Accessing nested values
print(countries["India"]["capital"])
print(countries["India"]["states"]["UP"])
```

```python
#Adding element to nested dictionary
countries["France"]={
    "capital":"Paris",
    "currency":"EUR",
    "states":{"IDF":"Paris","PAC":"Marseille"}
}
print(countries)
```

```python
#Update the nested dictionary
countries["France"]["currency"]="EURO"
print(countries)
```
