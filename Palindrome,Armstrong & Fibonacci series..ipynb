```python
(i)Write a python program to check if the given number is prime or not?
```

```python
num=int(input("enter a number:"))
if num > 1:
    for i in range(2,num):
        if (num % i) == 0:
            print(num, "is NOT a Prime Number")
            break
    else:
        print(num, "is a Prime Number")
else:
    print(num, "is NOT a Prime Number")
```

```python
(ii)Write a program to check for palindrome,armstrong & Fibonacci series.
```

```python
#PALINDROME
num=int(input("Enter any number"))
n=num
rev=0
while num>0:
    rev=(rev*10)+num%10  
    num=num//10          
if rev==n:
    print(f"Number {n} is palindrome")
else:
    print(f"Number {n} is not palindrome")
```

```python
#ARMSTRONG
num=int(input("enter a number:"))
num_str=str(num)
n=len(num_str)
sum=sum(int(number)**n for number in num_str)
if sum==num:
    print(num,"is an armstrong")
else:
    print(num,"is not an armstrong")
```

```python
#FIBONACCI
n = int(input("Enter number of terms: "))

a, b = 0, 1
print("Fibonacci Series:")

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
(iii)Create a python module that contains a function to calculate the factorial of a number using recursion.
```

```python
def fact(f):
    if f==0 or f==1:
        return 1
    else:
        return f*fact(f-1)
f=6
print(f"The factorial of {f} is {fact(f)}")
```
